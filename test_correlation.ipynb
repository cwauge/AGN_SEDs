```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.size'] = 22
plt.rcParams['axes.linewidth'] = 2
plt.rcParams['xtick.major.size'] = 3
plt.rcParams['xtick.major.width'] = 2
plt.rcParams['ytick.major.size'] = 3
plt.rcParams['ytick.major.width'] = 2
plt.rcParams['xtick.minor.size'] = 1.
plt.rcParams['xtick.minor.width'] = 0.5
plt.rcParams['ytick.minor.size'] = 1.
plt.rcParams['ytick.minor.width'] = 0.5
```

```python
Lx = np.linspace(39,50,1000)
FIR_const = np.ones(np.shape(Lx))
FIR_const[FIR_const == 1] = 43

mu, sigma = 43, 0.5 # mean and standard deviation
FIR_norm = np.random.normal(mu, sigma, 1000)
FIR_inc = np.linspace(41,46,1000)
FIR_inc2 = np.linspace(39,50,1000)
FIR_dec = np.linspace(44,39,1000)
```

```python
plt.figure(figsize=(8,8),facecolor='white')
plt.title('Constant FIR')
plt.plot(Lx, FIR_const - Lx,'.')
plt.xlabel('Lx')
plt.ylabel('FIR/Lx')
plt.xlim(40,46)
plt.ylim(-3,3)
plt.grid()
plt.show()

plt.figure(figsize=(8,8),facecolor='white')
plt.title('Normal FIR')
plt.plot(Lx, FIR_norm - Lx,'.')
plt.xlabel('Lx')
plt.ylabel('FIR/Lx')
plt.xlim(40,46)
plt.ylim(-3,3)
plt.grid()
plt.show()

plt.figure(figsize=(8,8),facecolor='white')
plt.title('Increasing FIR')
plt.plot(Lx, FIR_inc2 - Lx,'.')
plt.xlabel('Lx')
plt.ylabel('FIR/Lx')
plt.xlim(40,46)
plt.ylim(-3,3)
plt.grid()
plt.show()

plt.figure(figsize=(8,8),facecolor='white')
plt.title('Decreasing FIR')
plt.plot(Lx, FIR_dec - Lx,'.')
plt.xlabel('Lx')
plt.ylabel('FIR/Lx')
plt.xlim(40,46)
plt.ylim(-3,3)
plt.grid()
plt.show()
```
