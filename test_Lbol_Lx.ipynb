```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import ascii
```

```python
new = ascii.read('/Users/connor_auge/new_Lx_shape.txt')
old = ascii.read('/Users/connor_auge/old_Lx_shape.txt')


new_id = np.asarray(new['col1'])
new_Lx = np.asarray(new['col2'])
new_shape = np.asarray(new['col3'])

old_id = np.asarray(old['col1'])
old_Lx = np.asarray(old['col2'])
old_shape = np.asarray(old['col3'])
```

```python
plt.figure(figsize=(6,6),facecolor='white')
plt.plot(old_Lx,new_Lx,'.')
plt.plot(np.arange(43,46.5,0.5),np.arange(43,46.5,0.5),color='k')
plt.xlabel('old Lx')
plt.ylabel('new Lx')
plt.show()
```

```python
plt.figure(figsize=(6,6),facecolor='white')
plt.plot(old_shape,new_shape,'.')
plt.plot(np.arange(0,6,0.5),np.arange(0,6,0.5),color='k')
plt.xlabel('old shape')
plt.ylabel('new shape')
plt.show()
```
